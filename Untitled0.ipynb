```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df=pd.read_excel("amazon.xlsx")
```

```python
# KPI analysis
# Total analysis
df['TotalAmount'].sum()
```

```python
# total order

df['OrderID'].count()
```

```python
# average sales
df['TotalAmount'].mean()
```

```python
# Maximum profit
df['Quantity'].max()
```

```python
#REVENUE BY STATE
df.groupby('State')['TotalAmount'].sum()
```

```python
#REVENUE BY CATEGORY
df.groupby('Category')['TotalAmount'].sum()
```

```python
# TOP 5 CUSTOMERS
df.groupby('CustomerID')['TotalAmount'].sum()
```

```python
# ORDER BY PAYMENT MODE
df.groupby('PaymentMethod')['OrderID'].count()
```

```python
# visuliazation
df.groupby('State')['TotalAmount'].sum().plot(kind='bar')
plt.title('Revenue by State')
plt.xlabel('State')
plt.ylabel('Revenue')
plt.show()
```

```python
df.groupby(['CustomerID','CustomerName'])['TotalAmount'].sum().sort_values(ascending=False).head(5)
```

```python
# category wise sales
df.groupby('Category')['TotalAmount'].sum().sort_values(ascending=False).plot(kind='pie')
plt.title('Category wise sales')
plt.show()
```
